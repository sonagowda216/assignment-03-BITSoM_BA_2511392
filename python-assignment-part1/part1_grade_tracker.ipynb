# Part 1 — Student Grade Tracker

This notebook runs Tasks 1–4 from the assignment and captures visible outputs.

- Single-file submission: `part1_grade_tracker.py` is imported as a module.
- All tasks are executed below and outputs are embedded for submission verification.

````python
# Import module (kept minimal)
import importlib
import part1_grade_tracker as pg
importlib.reload(pg)
from pprint import pprint
import pandas as pd
````

````python
# Task 1 — Data Parsing & Profile Cleaning (run & show outputs)

raw_students = [
    {"name": " ayesha SHARMA ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": " ROHIT verma ", "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": " Priya Nair ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": " karan MEHTA ", "roll": "104", "marks_str": "40, 55, 38, 62, 58"},
    {"name": " Sneha pillai ", "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

students = pg.clean_student_data(raw_students)
print('\nReturned cleaned students list:')
from pprint import pprint
pprint(students)

# Also show as a table
try:
    display(pd.DataFrame(students))
except Exception:
    pass
````

````python
# Task 2 — Compute Results & Class Statistics (run & show tables)

students = pg.compute_student_results(students)
class_stats = pg.compute_class_statistics(students)

print('\nStudents (tabular):')
try:
    display(pd.DataFrame(students))
except Exception:
    from pprint import pprint
    pprint(students)

print('\nClass statistics:')
from pprint import pprint
pprint(class_stats)
````

````python
# Task 2 — Marks Analysis (loops & conditionals) with visible prints
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

print('\nSubjects with grades:')
pg.print_subjects_with_grades(subjects, marks)

print('\nMarks summary:')
summary = pg.compute_total_avg_high_low(subjects, marks)
from pprint import pprint
pprint(summary)
````

````python
# Interactive marks-entry (simulation) to show validation and updated average
simulation = [
    ("Biology", 82),
    ("Art", "abc"),  # invalid
    ("Geography", 74),
    ("done", None),
]
res = pg.interactive_marks_entry(subjects[:], marks[:], simulation=simulation)
print('\nInteractive entry result:')
from pprint import pprint
pprint(res)
````

````python
# Task 3 — Individual & Class Summary Reports (rendered)
from IPython.display import Markdown, display

for s in students:
    report = pg.generate_individual_report(s)
    display(Markdown('```
' + report + '\n```'))

class_report = pg.generate_class_summary_report(students, class_stats)
print('\nClass Summary Report:')
print(class_report)
````

````python
# Task 4 & Special Operations
pg.print_special_operations(students)
print('\nSubject-wise stats:')
from pprint import pprint
pprint(pg.get_marks_statistics_per_subject(students))

example_essay = (
    "This is the first sentence. Here is the second sentence! "
    "Is this the third sentence? Yes it is"
)
pg.task_4_string_utility(example_essay)
````
